Ejercicio 2 - Listas por comprensión

Usa comprensión de listas para:

a. Obtener los cuadrados de los primeros 10 números.

b. Filtrar los divisibles por 3 entre 1 y 30.

c. Generar una lista de pares (x, y) donde x < y y ambos estén en [1..5].

Define una función que devuelva los divisores de un número usando lista por comprensión.

divisores :: Int -> [Int]

Define la función divisoresPropios, crea una función que determine si un número es perfecto.

esPerfecto :: Int -> Bool


a. Obtener los cuadrados de los primeros 10 números


```haskell
cuadradosPrimeros10 :: [Int]
cuadradosPrimeros10 = [x^2 | x <- [1..10]]

cuadradosPrimeros10
```

b. Filtrar los divisibles por 3 entre 1 y 30


```haskell
divisiblesPor3 :: [Int]
divisiblesPor3 = [x | x <- [1..30], x `mod` 3 == 0]

divisiblesPor3
```

c. Generar una lista de pares (x, y) donde x < y y ambos estén en [1..5]


```haskell
paresXY :: [(Int, Int)]
paresXY = [(x, y) | x <- [1..5], y <- [1..5], x < y]

paresXY
```

Función divisores


```haskell
divisores :: Int -> [Int]
divisores n = [x | x <- [1..n], n `mod` x == 0]

divisores 28
```

Función divisoresPropios y función esPerfecto


```haskell
divisoresPropios :: Int -> [Int]
divisoresPropios n = [x | x <- [1..n-1], n `mod` x == 0]

divisoresPropios 28

esPerfecto :: Int -> Bool
esPerfecto n = sum (divisoresPropios n) == n

esPerfecto 28
esPerfecto 12
```
