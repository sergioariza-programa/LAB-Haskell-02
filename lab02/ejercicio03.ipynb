sumaLista :: [Int] -> Int

mayoresQue10 :: [Int] -> Int

productoLista :: [Int] -> Int

celsiusAFahrenheit :: [Double] -> [Double]

ajustarImpares :: [Int] -> [Int]

nombresLargos :: [String] -> [String]

soloNegativos :: [Int] -> [Int]

Funciones de orden superior


```haskell
sumaLista :: [Int] -> Int
sumaLista = foldr (+) 0

sumaLista [1,2,3,4]

mayoresQue10 :: [Int] -> Int
mayoresQue10 = length . filter (>10)

mayoresQue10 [3,10,11,12,20]

productoLista :: [Int] -> Int
productoLista = foldr (*) 1

productoLista [2,3,4]

celsiusAFahrenheit :: [Double] -> [Double]
celsiusAFahrenheit = map (\c -> c * 9 / 5 + 32)

celsiusAFahrenheit [0,12,20]

ajustarImpares :: [Int] -> [Int]
ajustarImpares = map (\x -> if odd x then x + 5 else x)

ajustarImpares [1,2,3,4,5,6]
```

Funciones usando filter


```haskell
nombresLargos :: [String] -> [String]
nombresLargos = filter (\nombre -> length nombre > 5)

nombresLargos ["Ana","Camilo","Valentina","Luis"]

soloNegativos :: [Int] -> [Int]
soloNegativos = filter (<0)

soloNegativos [-3,5,-1,0,8,-9]

edadesUniversitarios :: [Int] -> [Int]
edadesUniversitarios = filter (\edad -> edad >= 18 && edad <= 25)

edadesUniversitarios [16,18,20,25,27,30]
```

Funciones usando filter y map


```haskell
agregarIVA :: [Double] -> [Double]
agregarIVA = map (\x -> x * 1.19) . filter (>=1000)

agregarIVA [500,1000,1500,2000]

longitudesVocales :: [String] -> [Int]
longitudesVocales = map length . filter (\palabra -> not (null palabra) && head palabra `elem` "aeiouAEIOU")

longitudesVocales ["uva","casa","elefante","oso","perro"]

cuadradosPares :: [Int] -> [Int]
cuadradosPares = map (^2) . filter even

cuadradosPares [1..10]
```

Funciones usando any o all


```haskell
import Data.Char (isUpper)

hayNegativo :: [Int] -> Bool
hayNegativo = any (<0)

hayNegativo [4,8,-1,3]

todosPares :: [Int] -> Bool
todosPares = all even

todosPares [2,4,6,8]
todosPares [2,3,4]

nombresCorrectos :: [String] -> Bool
nombresCorrectos = all (\nombre -> not (null nombre) && isUpper (head nombre))

nombresCorrectos ["Ana","Luis","Pedro"]
nombresCorrectos ["ana","Luis","Pedro"]
```

Funciones usando takeWhile o dropWhile


```haskell
menoresQue100 :: [Int] -> [Int]
menoresQue100 = takeWhile (<100)

menoresQue100 [3,15,80,99,100,120,150]

sinCerosIniciales :: [Int] -> [Int]
sinCerosIniciales = dropWhile (==0)

sinCerosIniciales [0,0,0,3,4,0,5]
```
