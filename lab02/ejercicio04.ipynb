# Ejercicio 4

Ejercicio 4 - Funciones que reciben y retornan funciones

Define una función que recibe una función que transforma un dato y lo aplica a una lista de datos.

procesarLecturas :: (Double -> Double) -> [Double] -> [Double]

Implementa procesarLecturas

Ejemplo:

procesarLecturas celsiusAFaranheit [0,12,20]

Define una función que retorne un filtro personalizado.

crearFiltro :: (a -> Bool) -> ([a] -> [a])

Implementa crearFiltro

Ejemplo:

soloPares = crearFiltro even
soloPares [1..10] == [2,4,6,8,10]

Crea una función que genera verificadores de rangos.

enRango :: Int -> Int -> (Int -> Bool)

Implementa enRango

Ejemplo:

entre5y10 = enRango 5 10
entre5y10 7

Genera una función que eleva a una potencia dada.

potenciador :: Int -> (Int -> Int)

Implementa potenciador

Ejemplo:

potencia5 = potenciador 5
potencia5 7


procesarLecturas


```haskell
celsiusAFahrenheit :: Double -> Double
celsiusAFahrenheit c = c * 9 / 5 + 32

procesarLecturas :: (Double -> Double) -> [Double] -> [Double]
procesarLecturas f xs = map f xs

procesarLecturas celsiusAFahrenheit [0,12,20]
```

crearFiltro


```haskell
crearFiltro :: (a -> Bool) -> ([a] -> [a])
crearFiltro criterio = filter criterio

soloPares :: [Int] -> [Int]
soloPares = crearFiltro even

soloPares [1..10]
```

enRango


```haskell
enRango :: Int -> Int -> (Int -> Bool)
enRango minimo maximo = \x -> x >= minimo && x <= maximo

entre5y10 :: Int -> Bool
entre5y10 = enRango 5 10

entre5y10 7
entre5y10 12
```

potenciador


```haskell
potenciador :: Int -> (Int -> Int)
potenciador n = \x -> x ^ n

potencia5 :: Int -> Int
potencia5 = potenciador 5

potencia5 7
```
